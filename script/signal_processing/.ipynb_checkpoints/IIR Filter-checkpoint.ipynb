```python
from scipy import signal
from scipy.signal import filter_design as fd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import pandas as pd


# [each sample is 40us apart and the key signal is 100Hz]

# Sample Frequency 25KHZ
Fs = 25e3
T  = 1/Fs

# Key signal frequency 100HZ = 6000RPM
f     = 100

# create an array of  points for 12-bit resolution
N=np.int32((Fs/f)*10)

t  = np.arange(0,(N*T),T)

Omega = 2*np.pi*f

# noise generator: following datasheet 

noise=(np.random.normal(0.9,0.1,N)/360)*4095

for i,k in enumerate(noise):
    noise[i]=k*np.random.choice([-1,+1])

# Input signal generator    
x  = np.int32(np.round((((Omega*t)%(2*np.pi))/(2*np.pi))*4095+noise))

x_ideal=np.int32(np.round((((Omega*t)%(2*np.pi))/(2*np.pi))*4095))

tInput_q=(15.0/360.0)*4095+noise
# keep x[] in the right range
for i, a in enumerate(x):
    if x[i]>=4095:
        x[i]=4095
    if x[i]<=0:
        x[i]=0

fig1 = plt.figure(figsize=(16,12),dpi=80)        
fig1.add_subplot(311)
plt.plot( x,'r-', linewidth=1)
plt.ylabel( 'ip value' )
plt.grid()

fig1.add_subplot(312)
plt.plot( np.abs(noise),'k-', linewidth=1)
plt.ylabel( 'Noise value' )
plt.grid()

fig1.add_subplot(313)
plt.plot( tInput_q,'b-', linewidth=1)
plt.xlabel( 'sample length' )
plt.ylabel( 'Quiescent value' )
plt.grid()

plt.subplots_adjust(hspace=0.6)
plt.show()

```

```python
# FIR filter
#For signal.lfilter
#a[0]*y[n] = b[0]*x[n] + b[1]*x[n-1] + ... + b[nb]*x[n-nb]- a[1]*y[n-1] - ... - a[na]*y[n-na]

# low pass filter 
numtaps = 3
f_cutoff = 0.1

Num_fir=signal.firwin(numtaps, f_cutoff)

w, h = signal.freqz(Num_fir,[1])


# Denominator =1 means FIR
yf = signal.lfilter(Num_fir,[1],x)

yf_q=signal.lfilter(Num_fir,[1],tInput_q)

fig1 = plt.figure(figsize=(16,12),dpi=80)        
fig1.add_subplot(411)
plt.plot( x,'r-', linewidth=1)
plt.ylabel( 'Motion value' )
plt.grid()

fig1.add_subplot(413)
plt.plot(tInput_q,'k-', linewidth=1)
plt.ylabel( 'Quiescent value' )
plt.grid()

fig1.add_subplot(412)
plt.plot( yf,'b-', linewidth=1)
plt.ylabel( 'Quiescent value' )
plt.grid()

fig1.add_subplot(414)
# Since 3-order FIR got 3 sample delay. Here just ignore the first 3 values
plt.plot( yf_q[2:],'b-', linewidth=1)
plt.xlabel( 'sample length' )
plt.ylabel( 'Quiescent value' )
plt.grid()

plt.subplots_adjust(hspace=0.6)
plt.show()

fig2 = plt.figure(2)
plt.title('Digital filter frequency response')
ax1 = fig2.add_subplot(111)

plt.semilogy(w, np.abs(h), 'b')

plt.ylabel('Amplitude (dB)', color='b')
plt.xlabel('Frequency (rad/sample)')

ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))

plt.plot(w, angles, 'g')
plt.ylabel('Angle (radians)', color='g')


plt.grid()
plt.axis('tight')
plt.show()


# the test result indicates that 3-order FIR can't perform well in this application
```

```python
# calculate the result manually and import reset error 

Input_signal=tInput_q
Output_Filtered_man  = [0]*len(tInput_q)

# filter execution
def filter(Input_signal,Filters,Reset_Error):
    a0 = Filters[0][0]
    a1 = Filters[0][1]
    a2 = Filters[0][2]
    b1 = Filters[1][1]
    b2 = Filters[1][2]
    rFlag=0
    rTimes=0
    print a0,a1,a2,b1,b2
    Output_Filtered=[0]*len(Input_signal)
    Output_Filtered[0]=Input_signal[0]
    Output_Filtered[1]=Input_signal[1]
    for n in range(2, len(Input_signal)):
        if rFlag==0:
            Output_Filtered[n] = a0*Input_signal[n] + a1*Input_signal[n-1] + a2*Input_signal[n-2] - b1*Output_Filtered[n-1] - b2*Output_Filtered[n-2]
        if (np.abs(Output_Filtered[n]-Input_signal[n]))>= Reset_Error:
            rFlag=1
            rTimes=2
        if rFlag==1:
            Output_Filtered[n]=Input_signal[n]
            rTimes=rTimes-1
        if rTimes==0:
            rFlag=0
        if Output_Filtered[n]<0:
            Output_Filtered[n]=0
        if Output_Filtered[n]>4095:
            Output_Filtered[n]=4095
        
    return Output_Filtered 

Output_Filtered_man=filter(Input_signal,Filters['butter'],100)

plt.subplot(111)
plt.plot( tInput_q,'r-', linewidth=2)
plt.plot( Output_Filtered_man,'y-', linewidth=2)
plt.xlabel( 'sample length' )
plt.ylabel( 'ip value' )
plt.grid()
plt.show()

#plt.subplot(212)

#plt.xlabel( 'sample length' )
#plt.ylabel( 'op value' )
#plt.grid()
#plt.show()
```

```python
# make degree into [-180,+180]
def unwrap_deg(deg):
    new_deg=0.0
    if deg>180.0:
        new_deg=deg-360.0
    if deg<(-180.0):
        new_deg=deg+360.0
    else:
        new_deg=deg
        
    return new_deg

Angle_Motion_data_original=pd.read_csv('C:/Linear & Angle/Test Data/Test Data Motion original.csv')

Angle_Motion_data_linearized=pd.read_csv('C:/Linear & Angle/Test Data/Test Data Motion linearized.csv')

time_Motion_origin=Angle_Motion_data_original['Time']
register_value_Motion_origin=Angle_Motion_data_original['Register Value']

register_value_Motion_filtered=filter(register_value_Motion_origin,Filters['butter'],100)

Angle_read_origin=[0.0]*len(register_value_Motion_origin)

for i in range(len(register_value_Motion_origin)):
    Angle_read_origin[i]=(register_value_Motion_origin[i]/4095.0)*360.0

Angle_read_filtered=[0.0]*len(register_value_Motion_filtered)
for i in range(len(register_value_Motion_filtered)):
    Angle_read_filtered[i]=(register_value_Motion_filtered[i]/4095.0)*360.0    
    
ref_value1=Angle_Motion_data_original['Optical Encoder']

# calculate the angle error and transform it into radians -Pi to +Pi

angleError_filtered=(ref_value1-Angle_read_filtered)
for i in range(len(angleError_filtered)):
    angleError_filtered[i]=unwrap_deg(angleError_filtered[i])
angleError_filtered=np.deg2rad(angleError_filtered)
#angleError_filtered=np.unwrap(angleError_filtered)

angleError_original=(ref_value1-Angle_read_origin)
angleError_original=np.deg2rad(angleError_original)
angleError_original=np.unwrap(angleError_original)




time_Motion_linearized=Angle_Motion_data_linearized['Time']
register_value_Motion_linearized=Angle_Motion_data_linearized['Register Value']
ref_value2=Angle_Motion_data_linearized['Optical Encoder']

fig9 = plt.figure(figsize=(16,12),dpi=80)
plt.title('A1335 register value in motion state')
plt.subplot(311)
plt.plot(time_Motion_origin,Angle_read_origin,'b-', linewidth=1)
plt.plot(time_Motion_origin,ref_value1,'r-', linewidth=1)
plt.xlabel( 'sample length' )
plt.ylabel('Original register reading')
plt.grid()

plt.subplot(312)
plt.plot(time_Motion_origin,angleError_original,'b-', linewidth=1)
#plt.plot(time_Motion_origin,angleError_filtered,'r-', linewidth=1)
plt.xlabel( 'sample length' )
plt.ylabel('Non filtered angle reading')
plt.grid()

plt.subplot(313)
#plt.plot(time_Motion_origin,angleError_original,'b-', linewidth=1)
plt.plot(time_Motion_origin,np.abs(angleError_filtered),'r-', linewidth=1)
plt.xlabel( 'sample length' )
plt.ylabel('Filtered angle reading')
#plt.ylim([0.03,0.065])
plt.grid()
plt.show()
```
